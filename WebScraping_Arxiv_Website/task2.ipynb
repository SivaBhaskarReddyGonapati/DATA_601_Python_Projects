Create a dataframe for each primary category (Databases,Graphics,Robotics,Emerging Technologies)

<b>Code to create dataframe for each category from stored CSV file data</b>

```python
#importing pandas to create dataframes
import pandas as pd

#loading the metadata from the CSV file
metadata_df = pd.read_csv('arxiv_metadata.csv')

#defining the categories
categories = ["cs.DB", "cs.GR", "cs.RO", "cs.ET"]

#creating a dictionary to store the dataframes for each primary category
category_dfs = {}

#looping through the categories and creating the dataframes
for category in categories:
    category_dfs[category] = metadata_df[metadata_df['primary_category'] == category]
```

<b>Displaying the first five lines of each category dataframe</b>

```python
#displaying the first 5 lines of each primary category dataframe
for category in categories:
    print(f"\nFirst 5 lines of {category} category:\n")
    display(category_dfs[category].head(5))
```
